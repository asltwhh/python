# 转义字符
- 用一个特色的方法表示出一系列不方便写出的内容，比如回车键，换行符，退格键
- 出现反斜杠要小心
- 不同的系统对于换行操作有不同的表示
    - windows:\n
    - linux：\r\n


```python
s = 'Let\'s go'  #中间的'需要使用转义字符表示\'
print(s)

# 或者使用双引号嵌套单引号使用
s1 = "Let's go"
print(s1)

s2 = "c:\\Use\\adjfdkjt"   # 双斜杠表示单斜杠
print(s2)

s3 = "sdjf\nhjds\njds"
print(s3)
```

# 常用的转义字符
- \(在行尾时)	续行符
- \\	反斜杠符号
- \'	单引号
- \"	双引号
- \a	响铃
- \b	退格(Backspace)
- \e	转义
- \000	空
- \n	换行
- \v	纵向制表符
- \t	横向制表符
- \r	回车
- \f	换页
- \oyy	八进制数yy代表的字符，例如：\o12代表换行
- \xyy	十进制数yy代表的字符，例如：\x0a代表换行
- \other	其它的字符以普通格式输出

```python
# 单个斜杠的用法，表示此行未结束，处于美观需要在下一行继续
s = "这一句话并没有结束\
未完待续"
print(s)

#响铃-----就是一个小方框
s1 = "你好吗\a朋友"
print(s1)
```

# 传统格式化
- 使用%进行格式化
- %（百分号）也叫占位符
    - %s    字符串 (采用str()的显示)
    - %r    字符串 (采用repr()的显示)
    - %c    单个字符
    - %b    二进制整数
    - %d    十进制整数
    - %u    无符号整数
    - %o    八进制整数
    - %x    十六进制整数
    - %e    指数 (基底写为e)
    - %E    指数 (基底写为E)
    - %f    浮点数 %.3f 表示该浮点数包含三位小数
    - %F    浮点数，与上相同
    - %g    指数(e)或浮点数 (根据显示长度)
    - %G    指数(E)或浮点数 (根据显示长度)
    - %%    字符"%"
- 格式字符前出现的整数表示占位符所占位置的宽度
- 格式字符前边出现- 表示左对齐，+ 表示右对齐
- 0位数不足用“0”补齐
- width表示宽度
- pricision表示精度

```python
# 用%s表示一个简单的字符串
s = "I love %s xiaojing"
print(s%"wang")  #替换
```

```python
print("I love %s"%"li xiaojing")
# %s比较奇特，但是一般占位符只能被同类型的替换  特例
print("I love %s"%100)

#但是%d时只能接受数字
print("i am %d years old"%10)
#以下打印会报错
#print("i am %d years old"%'10')

#当需要格式化的参数多于一个，则需要使用小括号
print("i'm %f weight,%f height"%(50.2,1.6)) 
print("i'm %.2f weight,%.1f height"%(50.2,1.6))    #由于上面格式存在很多的小数点，则需要对其位数进行限制
```

# format格式化
- 使用函数形式代替之前的格式化%

```python
# 不用指定位置，按顺序读取

#方式1
s = "{} {}!"
print(s.format("hello","world"))

#方式2
s = "{} {}!".format("hello","world")
print(s)

#设定指定位置
s = "{0} {1}!".format("hello","world")
print(s)

s = "{1} {1}!".format("hello","world")
print(s)

s = "i love {0} and {0} loves me".format("wangxiaojing")
print(s)

# s = "i love {} and {} loves me".format("wangxiaojing")  报错
# print(s)

#使用命名参数
s= "我们是{schoolname}，我们的网址是{url},{teacher}最帅"
s.format(schoolname="wuhanligong",url="www.whut.com",teacher="quedashun")#不会改变s的值
s = s.format(schoolname="wuhanligong",url="www.whut.com",teacher="quedashun")
print(s)
```

```python
# 通过字典设置参数，需要解包
#使用命名参数
s= "我们是{schoolname}，我们的网址是{url},{teacher}最帅"
s_dict={"schoolname":"wuhanligong",\
        "url":"www.whut.com",\
        "teacher":"quedashun"}
s = s.format(**s_dict) #加上**表示解包字典
print(s)
```

```python
#对数字的格式化
# :后面带填充的字符，只能有一个
# + 表示正数显示，- 表示负数显示
# b,d,o,x表示十、二、八、十六进制数
# ^表示居中， <表示左对齐, >表示右对齐
# 大括号转义：{{表示{，}}表示}
s = "Wu huanhuan is {{{:.2f}}}m height,{{{:.2f}}}KG weight"   
print(s.format(1.61,50.2))
```
