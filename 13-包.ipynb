# 1 模块
- 一个模块就是一个包含python代码的文件，后缀名是.py，模块就是个python文件
- 为什用模块
    - 程序太大
    - 程序可以增加代码重复使用的方式
    - 当做命名空间使用，避免命名冲突
- 如何定义模块
    - 模块就是一个普通文件，代码直接书写
    - 不过根据模块的规范，最好在模块中编写以下内容：
        - 函数（单一功能）
        - 类（相似功能的组合，或者类似业务模块）
        - 测试代码
- 如何实用模块
     - 直接导入
     - 语法1：
             import module_name
             module_name.function_name
             module_name.class_name
     - 模块的命名和变量的命名相同，不允许以数字开头
     - 一旦模块的命名以数字开头，则可以借助importlib这个包导入
             import importlib
             p = importlib.import_module("01") # 导入了一个模块01，并将其赋值给了p
     - 语法2：导入模块的同时简化模块名
             import module_name as 别名
             别名.function_name
             别名.class_name
     - 语法3：明确给出导入该模块的那个方法或者类，不需要再使用模块名作为前缀
             from module_name import func_name,class_name
     - 语法4：相当于import module_name ,导入所有内容，不需要再使用模块名作为前缀
             from module_name import * 
- 对于模块中的print语句，可以通过下面的方式选择是否执行
         if __name__=='__main__':
             print("hdfsj......")
         - 这样写就可以保证当我们执行本模块文件时，Print语句得到执行，但是当该文件作为模块传入其他文件时不执行该语句
         - 建议一直使用此语句作为程序的入口

# 包
- 包是一种组织管理的方式，包里面存放的是模块
- 用于将模块包含在一起的文件夹就是包
- 自定义的结构
        包
            __init__.py文件  #包的标志文件
            模块1
            模块2
            子包（子文件夹）
                __init__.py文件
                子包模块1
                子包模块2
- 包的导入操作
    - import package_name
        - 直接导入一个包，只导入了__init__.py中的内容
        - （1）使用方式是:
                package_name.func_name
                package_name.class_name.func_name()
        - 此种方式的访问内容是
        - 查看D:\Workspaces\NN\vscode\包 pkg01和p07.py
        - （2）别名：import package_name as 别名
            - 只导入了__init__.py中的内容
            - 用法与上面（1）相同，只是名称有变化
            - 注意此种方法默认导入__init__.py内容
        - （3）import package.module 直接导入包的某一个模块，需要前缀
            - package_name.module_name.func_name()
            - 导入包内的某个具体的模块 
            - 详见p08.py
        - （4）import package.module as 别名
            - 使用方法等同于上面（3）
- from .. import ...
    - from package import module1,module2,module3,...
    - 此种方法不执行__init__.py文件的内容
            from pkg01 import p01
            p01.sayHello()
    - from package import *
        - 导入当前包中的__init__文件中的所有函数和类，比较恶心
        - 使用方法：不需要前缀
                func_name()
                class_name.func_name()
                class_name.var
        - 案例p09.py # 注意它导入的内容
- from package.module import * 
    - 导入包中指定的模块的所有内容
    - 使用方法： 不需要前缀
            func_name()
            class_name.func_name()
            class_name.var 
    - 案例见p10.py
- 在开发环境中经常会使用到其他模块的内容，可以在当前包中直接导入该模块的内容
    - import 完整的包或者模块的路径
- ‘__all__’的用法
    - 可以规定在使用from package import *的时候需要导入的内容
    - __init__.py中如果文件为空，或者没有__all__,则只可以将__init__中的内容导入
    - __init__.py设置了__all__的值，则按照__all__指定的子包或者模块进行导入，不会载入__init__中的其他内容
    - __all__ = ['module1','module2','child_package1',...]
    - 案例pkg02,p11.py
- 命名空间
    - 用于区分不同位置不同功能名称相同的函数或者变量的一个特定前缀
    - 作用是防止命名冲突
    - 比如：下面的setName()函数分布在不同的命名空间内
            setName()
            Student.setName()
            Dog.setName()
